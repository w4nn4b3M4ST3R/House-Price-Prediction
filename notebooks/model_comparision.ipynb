# üè† House Price Prediction üíµ

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

### üîÉ Loading and processing data

```python
train_df = pd.read_csv('../data/train.csv',index_col='Id')
train_df
```

```python
train_df.columns
```

```python
features = ['LotArea','YearBuilt','1stFlrSF','1stFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
```

```python
X = train_df[features]
X
```

```python
Y = train_df['SalePrice']
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
print(X_train,Y_train)
```

```python
print(X_test,Y_test)
```

### üìñ Training

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(random_state=1)
dt_model.fit(X_train,Y_train)
```

```python
dt_predict_result = dt_model.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train,Y_train)
```

```python
rf_predict_result = rf_model.predict(X_test)
```

```python
gb_model = GradientBoostingRegressor(random_state=1)
gb_model.fit(X_train,Y_train)
```

```python
gb_predict_result = gb_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
dt_mae = mean_absolute_error(Y_test, dt_predict_result)
dt_mse = mean_squared_error(Y_test, dt_predict_result)
dt_r2 = r2_score(Y_test, dt_predict_result)
```

```python
rf_mae = mean_absolute_error(Y_test, rf_predict_result)
rf_mse = mean_squared_error(Y_test, rf_predict_result)
rf_r2 = r2_score(Y_test, rf_predict_result)
```

```python
gb_mae = mean_absolute_error(Y_test, gb_predict_result)
gb_mse = mean_squared_error(Y_test, gb_predict_result)
gb_r2 = r2_score(Y_test, gb_predict_result)
```

```python
print('======================================  üå≥ DecisionTree Model  ==========================================')
print("Mean Absolute Error:", dt_mae )
print("Mean Squared Error:", dt_mse)
print("R2 Score:", dt_r2)
```

```python
print('======================================  üå≥üå≥üå≥ RandomForest Model  ==========================================')
print("Mean Absolute Error:", rf_mae)
print("Mean Squared Error:", rf_mse)
print("R2 Score:", rf_r2)
```

```python
print('======================================  üîç Gradient Boosting Model  ==========================================')
print("Mean Absolute Error:", gb_mae)
print("Mean Squared Error:", gb_mse)
print("R2 Score:", gb_r2)
```

```python
from sklearn.linear_model import LinearRegression
x_dt_vals = np.linspace(min(Y_test), max(Y_test), 1000)
y_dt_vals = LinearRegression().fit(np.array(Y_test).reshape(-1,1), np.array(dt_predict_result).reshape(-1,1)).predict(x_dt_vals.reshape(-1,1))

x_rf_vals = np.linspace(min(Y_test), max(Y_test), 1000)
y_rf_vals = LinearRegression().fit(np.array(Y_test).reshape(-1,1), np.array(rf_predict_result).reshape(-1,1)).predict(x_rf_vals.reshape(-1,1))

x_gb_vals = np.linspace(min(Y_test), max(Y_test), 1000)
y_gb_vals = LinearRegression().fit(np.array(Y_test).reshape(-1,1), np.array(gb_predict_result).reshape(-1,1)).predict(x_gb_vals.reshape(-1,1))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(Y_test,dt_predict_result,alpha=0.6,color = 'yellow',label = 'prediction',edgecolor='k')
plt.plot(x_dt_vals,y_dt_vals,'--',linewidth=2, color='blue', label='Regression Line')
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)],'--', color='red', linewidth=2, label='Perfect prediction line',)
plt.xlabel('Reality',fontsize=12)
plt.ylabel('Prediction',fontsize=12)
plt.title('Decision Tree Model ',fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()

```

```python
plt.figure(figsize=(6,6))
plt.scatter(Y_test,rf_predict_result,alpha=0.6,color = 'yellow',label = 'prediction',edgecolor='k')
plt.plot(x_rf_vals,y_rf_vals,'--',linewidth=2, color='blue', label='Regression Line')
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)],'--', color='red', linewidth=2, label='Perfect prediction line',)
plt.xlabel('Reality',fontsize=12)
plt.ylabel('Prediction',fontsize=12)
plt.title('Random Forest Model ',fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(Y_test,gb_predict_result,alpha=0.6,color = 'yellow',label = 'prediction',edgecolor='k')
plt.plot(x_gb_vals,y_gb_vals,'--',linewidth=2, color='blue', label='Regression Line')
plt.plot([min(Y_test), max(Y_test)],[min(Y_test), max(Y_test)],'--', color='red', linewidth=2, label='Perfect prediction line',)
plt.xlabel('Reality',fontsize=12)
plt.ylabel('Prediction',fontsize=12)
plt.title('Gradient Boosting Model ',fontsize=14)
plt.legend()
plt.grid(True)
plt.tight_layout()
```
